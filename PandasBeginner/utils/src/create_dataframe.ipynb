Pandas入門で使うデータを作成するためのライブラリ

## I/O

- Input: 
  - None
- Output: 
  - score_sheet.csv
  - score_sheet.xlsx
  - score_sheet_non_header.csv
  - score_sheet_non_header.xlsx

```python
# ライブラリのインポート
import random

import pandas as pd

random.seed(0)
```

```python
data_dir = '../data/'
```

```python
CLASS = ['A', 'B', 'C', 'D', 'E']
```

```python
ids = [str(i).zfill(5) for i in range(1, 101)]
class_ = [random.choice(CLASS) for _ in range(100)]
japanese = [random.randint(0, 100) for _ in range(100)]
math = [random.randint(0, 100) for _ in range(100)]
english = [random.randint(0, 100) for _ in range(100)]
```

```python
df = pd.DataFrame({'id': ids,
                   'class': class_,
                   'japanese': japanese,
                   'math': math,
                   'english': english})
```

```python
df
```

```python
df.to_csv(data_dir + 'score_sheet.csv', index=None, encoding='utf-8-sig')
df.to_excel(data_dir + 'score_sheet.xlsx', index=None, encoding='utf-8-sig')
```

```python
tmp = pd.DataFrame(df.columns).T
df2 = df.copy()
df2.columns = tmp.columns
df2 = pd.concat([tmp, df2]).reset_index(drop=True)
df2
```

```python
df2.to_csv(data_dir + 'score_sheet_non_header.csv', index=None, encoding='utf-8-sig')
df2.to_excel(data_dir + 'score_sheet_non_header.xlsx', index=None, encoding='utf-8-sig')
```
