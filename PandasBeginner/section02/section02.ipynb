## Conda仮想環境の作成

まずはターミナルを立ち上げて、ダウンロードしたフォルダまで移動してあげます。

```
$ cd <ダウンロードしたフォルダ>
```

<br>

もしソースコードをデスクトップに置いたなら、以下のようになりますね。
```
$ cd Desktop/PandasBeginner
```

<br>

もしターミナル操作に慣れていなければ、cdを書いたあとにフォルダをドラック&ドロップすれば勝手にPATHを表示してくれます！

いずれかの方法で、いまダウンロードしてきたフォルダまで移動してあげましょう。

<br>

`PandasBeginner`に移動したら、さっそくCondaの仮想環境を作成していきたいと思います。

Condaコマンドを使って仮想環境を作成するためには、以下のコマンドを入力します。

```
$ conda create -n <環境名> python=<Python Version>
```

<br>

今回は`pandas_env`という環境名、Pythonバージョンは`3.8`を使っていきたいと思います。

つまり、入力すべきコマンドは次のようになります。

```
$ conda create -n pandas_env python=3.8
```

<br>

これでEnterキーを入力しましょう。途中でy/nを聞かれますが、すべてy(yes)で答えてしまって大丈夫です。

コマンドの実行が完了すると`pandas_env`を作成できているかと思います。

*※`conda info -e`を実行すると、作成済みのConda環境一覧を確認できます！*

<br>

最後にいま作成した仮想環境に切り替えておきましょう。

そのためには、以下のコマンドを入力します。

```
$ conda activate pandas_env
```

これで作成した仮想環境に切り替えできました！

## ライブラリのインストール

作成した`pandas_env`に切り替わっていることが確認できたら、ライブラリをインストールしていきます。

今回はとりあえず、Jupyter NotebookでPandasが使えれば良いので、以下のコマンドを入力してライブラリをインストールしていきます。

```
$ conda install pandas jupyter -y
```

<br>

ライブラリのインストールが完了したら、Jupyter Notebookを使ってみたいと思います。

ターミナルで`jupyter notebook`と入力して、Jupyter Notebookを立ち上げてみましょう！

<br>

おそらくJupyterのトップページにアクセスできるはずです！



## 拡張機能の導入

最後に拡張機能を導入していきましょう！

拡張機能を使うためには、`jupyter_contrib_nbextensions`をインストールする必要があります。

ですが本テキストを作成している時点(=2021/5/5)で、最新のJupyter Notebookに対応する拡張機能は、Condaコマンドでインストールできないようです。

<br>

なので`conda`ではなく`pip`を使ってインストールしていきましょう。

以下のコマンドを入力して`jupyter_contrib_nbextensions`をインストールします。

```
$ pip install jupyter_contrib_nbextensions
```

<br>

インストールが完了したら、Jupyter Notebookを立ち上げましょう。

```
$ jupyter notebook
```

Jupyterのトップ画面にアクセスすると、新しく「Nbextensions」というタブが追加されているはずです。



あとは使っていきたい拡張機能にチェックを入れるだけです。

拡張機能はたくさんありますが、今回は以下の4つを使っていきたいと思います。

- Table of Contents : 目次を追加する
- Variable Inspector : 変数の一覧を表示する
- Hinterland : 入力補完を使えるようにする
- Codefolding : セル内のコードを折り畳めるようにする

<br>

基本的にはチェックを入れるだけで機能します。

`Hinterland`はVSCodeでMicrosoftが提供している「Python拡張機能」のように動くので非常に便利です！

ぜひコードを書いて試してみましょう！
